```python
import pandas as pd

adult_census = pd.read_csv('/content/drive/MyDrive/Supervised ML/Working with numerical & categorical data/US-Adult-Census.csv')
print(adult_census.head())
```

```python
print(adult_census.dtypes)
```

```python
adult_census = adult_census.drop('education-num', axis=1)
```

Select only the numerical features

```python
numerical_cols = adult_census.select_dtypes(include=['int64']).columns
data, target = adult_census.drop(columns='class'), adult_census['class']
data_numerical = data[numerical_cols]
```

### K-Fold Cross-Validation

```python
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import make_pipeline
from sklearn.model_selection import cross_validate
```

```python
model_pip = make_pipeline(StandardScaler(), LogisticRegression(solver='lbfgs'))
cv_results = cross_validate(model_pip, data_numerical, target, cv=5)
print(cv_results)
```

extract the test scores from the cv_result dictionary and compute the mean accuracy
and the variation of the accuracy across folds.

```python
scores = cv_results['test_score']

print(f"The mean cross-validation accuracy is: {scores.mean():.3f} +/- {scores.std():.3f}")
```
